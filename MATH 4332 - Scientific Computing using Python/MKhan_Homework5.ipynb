```python
#Question 1
from string import ascii_lowercase
alphabetList = []
for i in range(1, len(ascii_lowercase) + 1):
  temp = ascii_lowercase[i - 1] * i
  alphabetList.append(temp)
print(alphabetList[0:12])
print(alphabetList[12:20])
print(alphabetList[20:])
```

```python
# Question 2
number = int(input("Enter a number: "))
factors = []
for i in range(1, number + 1):
  if number % i == 0:
    factors.append(i)
print("The factors of", number, "are:", factors)
```

```python
# Question 3
number_list = []
for i in range(11):
  number_list.append(1)
  number_list.extend([0] * i)
number_list.append(1)
print(number_list[:34])
print(number_list[34:])
```

```python
# Question 4
import random
random_nums = []
for i in range(100):
  random_nums.append(random.randint(0,1))
longest_seq = 0
current_seq = 0
for num in random_nums:
  if num == 0:
    current_seq += 1
  else:
    if current_seq > longest_seq:
      longest_seq = current_seq
    current_seq = 0
print("List of random numbers from 0-1:", random_nums[:50])
print(random_nums[50:])
print("Longest sequence of zeros in above list:", longest_seq)
```

```python
# Question 5
items = [1,1,2,3,4,3,0,0]
unique_items = []
for n in items:
  if n not in unique_items:
    unique_items.append(n)
print("List before removal of duplicates:", items)
print("List after removal of duplicates:", unique_items)
```
