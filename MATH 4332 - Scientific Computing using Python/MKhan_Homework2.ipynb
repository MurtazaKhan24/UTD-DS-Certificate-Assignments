```python
# Question 1
num = int(input("Enter a whole number of seconds: "))

print(num, "seconds is", num // 60, "minutes and", num % 60, "seconds.")
```

```python
# Question 2a
import math
num = int(input("Enter an integer: "))

print(num, "! is ", math.factorial(num), sep='')

# Question 2b
num = float(input("Enter an angle in degrees: "))
print("The sine of", num, "is", math.sin(num), "degrees.")
```

```python
# Question 3
num = float(input("Enter a temperature: "))
unit = input("Is it Celsius(0) or Farenheit(1): ")

if (unit == "0"):
  print(num, "degrees celsius is", (9/5)*num + 32, "degrees farenheit.")
else:
    print(num, "degrees farenheit is", (5/9)*(num - 32), "degrees celsius.")
```

```python
# Question 4
num = int(input("Enter the number of credits you've taken so far: "))

if(num >= 84):
  print("You are a senior")
elif(num >= 54):
  print("You are a junior")
elif(num >= 24):
  print("You are a sophomore")
else:
  print("You are a freshman")
```

```python
# Question 5
year = int(input("Enter a year: "))

if (year % 4 == 0):
  if(year % 100 == 0):
    if(year % 400 == 0):
      print(year, "is a leap year.")
    else:
      print(year, "is not a leap year.")
  else:
    print(year, "is a leap year.")
else:
  print(year, "is not a leap year.")
```
