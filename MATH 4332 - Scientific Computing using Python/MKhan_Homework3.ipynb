```python
# Question 1
scores = []
isGreater = False
for i in range(10):
  score = float(input("Enter a test score: "))
  scores.append(score)
  if score > 100:
    isGreater = True

scores.sort()
lowest = scores[0]
second_lowest = scores[1]
# part a
print("The lowest test score was:", lowest)
print("The second lowest test score was:", second_lowest)

# part b
if (isGreater):
  print("You entered a score greater than 100.")

# part c
scores.remove(lowest)
scores.remove(second_lowest)
avg_scores = sum(scores) / len(scores)
print("The mean scores was:", avg_scores)
```

```python
6# Question 2
import random
score = 0
for i in range(5):
  num = random.randint(1,10)
  guess = int(input("Enter a number to guess (1-10): "))
  if (guess == num):
    score += 10
    print("You guessed correctly!")
  else:
    score -= 1
    print("You guessed incorrectly.")
print("Game Over. Final Score:", score)
```

```python
6# Question 3
num_non_perfect_cubes = 0
count = 10
for i in range(1,1001):
  num = round(i ** (1/3))
  if num ** 3 != i:
    if count > 0:
      print(i, ",", sep='', end='')
      count -= 1
    num_non_perfect_cubes += 1
print("\nNumber of non perfect cubes from 1-1000:", num_non_perfect_cubes)
```

```python
# Question 5
import math
n = int(input("Enter an integer: "))
sum = 0
for i in range(1, n + 1):
  sum += (1/i)
sum -= math.log(n)
print("Sum is:", sum)
```
