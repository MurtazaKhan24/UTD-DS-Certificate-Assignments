```python
# Question 1
def weight_conversion(kg):
  return kg * 2.2

print("75 kilograms is", weight_conversion(75), "pounds.")
```

```python
# Question 2
x = float(input("Enter your first number: "))
y = float(input("Enter second number: "))
z = float(input("Enter third number: "))
print("The total is", x + y + z)
print("The average is", (x + y + z) / 2)
```

```python
# Question 3
price = float(input("Enter total price for meal: "))
percent = float(input("Enter tip amount in percent: "))
tip_amount = price * (percent / 100)
print("The total tip amount is:", tip_amount)
print("The total bill is:", price + tip_amount)
```

```python
# Question 4
for i in range(100,0,-2):
  print(i, end="," if i > 2 else ".")
```

```python
# Question 5
x = int(input("How many stars in the first line? "))
for i in range(x,0,-1):
  for j in range(i,0,-1):
    print("*", end="")
  print("")
```
