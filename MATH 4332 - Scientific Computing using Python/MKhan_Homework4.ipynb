```python
# Question 1
num_of_emails = int(input("How many emails do you want to enter?: "))
is_professor_email = False
for i in range(num_of_emails):
  email = input("Enter an email: ")
  if "@prof.college.edu" in email:
    is_professor_email = True
if is_professor_email:
  print("Some professor emails were entered")
else:
  print("All email addresses were student addresses")
```

```python
# Question 2
word = input("Enter a word: ")
result = ""
for i in range(len(word)):
  if i % 2 == 1:
    result += word[i].upper()
  else:
    result += word[i]
print(result)
```
